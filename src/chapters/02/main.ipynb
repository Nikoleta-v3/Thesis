**Fingerprinting** is a technique for generating a functional signature for a strategy. There are two types of fingerprints covered in this thesis. Namely:

- Fingerprinting 
- Transitive fingerprint

The Axelrod-Python library contains both these fingerprints, and with a few lines of code we can retrieve fingerprints for specific strategies.

**Fingerprinting for Pavlov (or WinStayLoseShift)**

```python
import axelrod as axl
```

```python
axl.seed(0)
strategy = axl.WinStayLoseShift
probe = axl.TitForTat
af = axl.AshlockFingerprint(strategy, probe)
data = af.fingerprint(turns=50, repetitions=2, step=0.01, progress_bar=False)
```

```python
p = af.plot(cmap='viridis')
```

```python
p.savefig('img/Win-Stay_Lose-Shift.pdf', bbox_inches='tight')
```

**Transitive Fingerprinting for Tit For Tat**

```python
axl.seed(0)
player = axl.TitForTat()
tf = axl.TransitiveFingerprint(player)
data = tf.fingerprint(turns=40)
```

```python
p = tf.plot()
```

```python
p.savefig('img/Tit_for_Tat_fingerprint.pdf', bbox_inches='tight')
```
